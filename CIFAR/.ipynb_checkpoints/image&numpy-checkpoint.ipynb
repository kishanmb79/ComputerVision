```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# this is a image function in python
from PIL import Image
```

```python
pic  = Image.open('DATA/00-puppy.jpg')
```

```python
pic
```

```python
type(pic)
```

```python
# here we are making the image readable by numpy 
pic_arr = np.asarray(pic)
```

```python
# this is of the type numpy
type(pic_arr)
```

```python
# this image has widthx height and 3 channel (RGB)
pic_arr.shape 
```

```python
plt.imshow(pic_arr)
```

```python
# lets make Green and blue channel Zero 
# and keep only red channel 
pic_red = pic_arr.copy()
#  RGB   R=0  G=1 B=2
# Red channel values  between 0-255
pic_red =  pic_red[ : , : , 0 ]
plt.imshow(pic_red,cmap='gray')
```

```python
# this is closer to green color in the image 
# closer to white means more green as it is in gray scale 
#  
pic_green = pic_arr.copy()
pic_green =  pic_green[ : , : , 1 ]
plt.imshow(pic_green,cmap='gray')
```

```python
# here O mean no blue color value 
# 255 mean full bue color value 
pic_blue  = pic_arr.copy()
pic_blue  = pic_blue[:,:,2]
plt.imshow(pic_blue,cmap='gray')
```

```python
# now let paly with this color 
```

```python

pic_test  = pic_arr.copy()
pic_test  = pic_test[:,:,:]
plt.imshow(pic_test,cmap='gray')
```

```python
pic_test.shape
```

```python
# let make green channel value zero 0
# RGB  G=1
pic_test[:,:,1] = 0
```

```python
# image without the color green  
plt.imshow(pic_test)
```

```python
# note that the color chanel green is there 
# it is just that the value is zero
print(pic_test.shape)
```

```python
# now lets take out blue 
# image only with red channel values 
pic_test[:,:,2] = 0
plt.imshow(pic_test)
# shape of all 3 channel 
print('shape of all 3 channel ',pic_test.shape)
# shape of only red channel
print('shape of only red channel ',pic_test[:,:,0].shape)
```
