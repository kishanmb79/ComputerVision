```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
img  = cv2.imread('DATA/rainbow.jpg',0)
plt.imshow(img, cmap='gray')
```

```python
# any value below 127 will be shifed to 0 and above 127 will be shifted to 255
ret, thresh_1 =  cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
# here ret is the cut off value and ththresh_1 is the image 
print(ret,' = ret')
print(thresh_1,' = thresh_1')
```

```python
plt.imshow(thresh_1,cmap='gray')
```

```python
img  = cv2.imread('DATA/crossword.jpg',0)
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th_1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th_1)
```

```python
# here we take in the image
# we take the max value 
# adaptiveMethod
# thresholdType 
# blockSize
# C
th_2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8) 
#
```

```python
show_pic(th_2)
```

```python
#
```

```python
blended = cv2.addWeighted(src1=th_1, alpha=0.6, src2=th_2, beta=0.4, gamma=0)
show_pic(blended)
```
