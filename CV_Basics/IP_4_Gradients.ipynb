```python
# image gradient is a directional change in the intensity or color in an image
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# here we are displaying the image
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax = plt.imshow(img, cmap='gray')
```

```python
img  = cv2.imread('DATA/sudoku.jpg',0)
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,dx=1,dy=0,ksize=5)
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,dx=0,dy=1,ksize=5)
display_img(sobely)
```

```python
# use Laplace operator
```

```python
laplaican  = cv2.Laplacian(img,cv2.CV_64F)
display_img(laplaican)
```

```python
# now lets try blending the image
```

```python
blended  = cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=0)
display_img(blended)
```

```python
ret, th_1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
display_img(th_1)
```

```python
#
```

```python
kernel  = np.ones((4,4),np.uint8)
gradient = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
display_img(gradient)
```
